| name | weight | height | BMI |

| Max | 110 | 70 | 15.78 |
| Oliver | 115 | 71 | 16.04 |
| Tom | 120 | 72 | 16.27 |
| Harry |125 | 73 | 16.49 |
| Jasper | 130 | 74 | 16.69 |
| Finn | 135 | 75 | 16.87 |
| Felix | 140 | 76 | 17.04 |
| Asher | 145 | 77 | 17.19 |

**a:** Generate a series with the names of all the people. Assign it to the variable `name`.

```python
# generating series with name of people and assigning it to variable 'name'

data = pd.DataFrame({ 'name': ['Max', 'Oliver', 'Tom', 'Harry', 'Jasper', 'Finn', 'Felix', 'Asher'],
                      'weight': [110,115,120,125,130,135,140,145],
                      'height': [70,71,72,73,74,75,75,77],
                      'BMI': [15.78,16.04,16.27,16.49,16.69,16.87,17.04,17.19]})

name = pd.Series(data['name'],name='name')
print(name)
```

**b:** Generate a series that contains the weight of all the people. Assign it to `weights`. (You are required to use a python function to do this. Hardcoding all values manually is not allowed).

```python
# Please write your code below
weights = pd.Series(data['weight'],name='weights')
print(weights)
```

**c:** Generate a series that contains the height of all the people. Assign it to `heights`. (You are required to use a python function to do this. Hardcoding all values manually is not allowed).

```python
# Generating series that contains the heights of people

heights = pd.Series(data['height'],name='heights')
print(heights)
```

**d:** Using the above series generate a series that contains the BMI of each person. The formula for BMI is as follows:
$$ BMI = weight (lb) ÷ height^2 (inches) * 703 $$
Assume that the given weight is in lb and height is in inches. Assign it to the variable `BMI`.  Round your answers to 2 decimal places  (You are not allowed to hardcode).

```python
# Calculating BMI
BMI = pd.Series(round((weights/(heights**2))*703,2),name='BMI')
print(BMI)
```

**e:** Finally, use all the above created series to create the dataframe shown above. Assign it to `BMIDaa`.

```python
# Generating dataframe "BMIData"

BMIData = pd.DataFrame({'name':name,'weight':weights,'height':heights,'BMI':BMI})
#df.columns = [col.title() for col in df.columns]
print(BMIData)
```
