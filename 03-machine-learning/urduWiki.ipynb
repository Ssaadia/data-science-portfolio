**Q23. Search for all the English words in the articles. Return a set of all English words. Note: A Set doesn't contain duplicates.**

````python
### write code here...

path = "/content/drive/MyDrive/ML/PA1/Dataset/UrduWiki.csv"
df = pd.read_csv(path, encoding = "utf-8")

english_words = set()
for article in df['content']:
  english_words.update(re.findall(r'[a-zA-Z]+', article))

print(english_words)
````

**Q24: Remove all punctuation marks, digits and single English characters from the articles. Return a list of all the cleaned articles. Punctuation marks include [؛ ، ٫ ؟ ۔ ٪ \ / ! @ ( ) ° . - ; ″ ′ \ -].**

````python
### write code here...
cleaned_articles = []

punctuation_marks = r'[؛ ، ٫ ؟ ۔ ٪ \ / ! @ ( ) ° . - ; ″ ′ \ -]'
digits = r'\d'
english_chars = r'\b[a-zA-Z]\b'

for article in df['content']:
  cleaned_article = re.sub(punctuation_marks, '', article)
  cleaned_article = re.sub(digits,'',article)
  cleaned_article = re.sub(english_chars,'',article)
  cleaned_articles.append(cleaned_article)

for article_text in cleaned_articles[:5]:
  print(article_text)
````

**Q25: Replace all the digits with Urdu digits, remove English words and remove punctuation marks. Urdu digits: [۰ ۱ ۲ ۳ ۴ ۵ ۶ ۷ ۸ ۹]**

````python
### write code here...

replace_digits = str.maketrans('0123456789', '۰۱۲۳۴۵۶۷۸۹')
punctuation_marks = r'[؛ ، ٫ ؟ ۔ ٪ \ / ! @ ( ) ° . - ; ″ ′ \ -]'
english_chars = r'\b[a-zA-Z]\b'

cleaned_articles = []

for article in df['content']:
  cleaned_article = article.translate(replace_digits)
  cleaned_article = re.sub(punctuation_marks, '', cleaned_article)
  cleaned_article = re.sub(english_chars,'',cleaned_article)
  cleaned_articles.append(cleaned_article)

for article_text in cleaned_articles[:5]:
  print(article_text)
````

**Q26: Return the top 10 most frequently used words starting with(م) in all the articles.**

````python
### write code here...

word_counter = Counter()
for article in df['content']:

  text = re.sub(r'[^\u0600-\u06FF\s]', '', article)
  words = text.split()

  for word in words:
    if word.startswith('م'):
      word_counter[word] += 1

most_frequent_words = word_counter.most_common(10)
print(most_frequent_words)
````
